```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
imported_df = pd.read_excel('data/precipitation/toulouse_excel.xlsx',usecols="B,H,AM")
```

```python
imported_df
```

```python
from dateutil import parser

for i in range(len(imported_df['Date'])):
    imported_df['Date'][i] = parser.parse(imported_df['Date'][i])
```

```python
df = imported_df.sort_values(by=['Date'])
df = df.rename(columns={"Température": "Temp", "Précipitations dans la dernière heure":"Precipitation"})
```

```python
df['Date']
```

```python
year_index = {}

for i in range(len(df['Date'])):
    if str(df['Date'].iloc(0)[i].year) not in year_index.keys():
        year_index[str(df['Date'].iloc(0)[i].year)] = [i]
    else:
        year_index[str(df['Date'].iloc(0)[i].year)].append(i)
```
